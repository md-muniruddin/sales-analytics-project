```python
## get the city name which has count more than 100
geo_city_count=(
    df.groupby('geolocation_city')
    .agg(
        total=('geolocation_city', 'count'),
    )
    .sort_values(by='total', ascending=False)
    .reset_index()
)
df_geo_name=geo_city_count[geo_city_count['total'] >= 500]
most_names_1=df_geo_name['geolocation_city']
```

```python
df_geo_name
```

```python
from deep_translator import GoogleTranslator
eng_name={}
for name in most_names_1:
    eng=GoogleTranslator(source='auto', target='en').translate(name)
    eng_name[name] = eng
```

```python
eng_name
```

```python

df['geolocation_city_en']=df['geolocation_city'].map(eng_name).fillna(df['geolocation_city'])
```

```python
df['geolocation_city_en'] = df['geolocation_city'].str.replace('ã', 'a').str.replace('ç', 'c')
```

```python
import pandas as pd
RAW_PATH='D:/DA project/sales-analytics-project/data/raw/olist_order_items_dataset.csv'
```

```python
order_items = pd.read_csv(RAW_PATH, parse_dates=False)
order_items.head(5)
```

```python
order_items.info()
```

```python
order_items.isnull().value_counts()
```

```python
order_items["shipping_limit_date"] = pd.to_datetime(
    order_items["shipping_limit_date"],
    errors="coerce"
)
```

```python
order_items["shipping_limit_date"].isna().mean()
```

```python
order_items["price"] = pd.to_numeric(order_items["price"], errors="coerce")
order_items["freight_value"] = pd.to_numeric(order_items["freight_value"], errors="coerce")
```

```python
order_items["is_zero_price"] = order_items["price"] == 0
order_items["is_negative_price"] = order_items["price"] < 0
order_items["is_high_freight"] = order_items["freight_value"] > order_items["price"]
```

```python
order_items[["is_zero_price", "is_negative_price", "is_high_freight"]].mean()
```
